## Some help, if you need it

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
import re

definition = {
    "a": "Bag of Words (binary)",
    "b": "Bag of Words (Term-Frequency) without normalization",
    "c": "Bigrams of Words (Term-Frequency) without normalization",
    "d": "Trigram of Characters (Term-Frequency) without normalization",
    "e": "TF-IDF based on words with 'l2' normalization",
    "f": "LSA based on TF-IDF of words (50 singular values)",
    "g": "LSA based on TF-IDF of words (100 singular values)"
}
# !pip install pandas
```

## Load de dataset

```python
filename = "EXIST2024_EN_examples.csv"

# COMPLETE
df = pd.read_csv(filename, sep='\t')
df.head(5)
```

## Preprocess dataset

```python
import re

web_re = re.compile(r"https?:\/\/[^\s]+", re.U)

def preprocess(text):
    # COMPLETE
    text = re.sub(web_re, '', text)
    text = text.lower()
    return text

# COMPLETE
df['text'] = df['text'].apply(preprocess)
df.head(5)
```

## Create the vectorizers and compute the representation of the tweets

```python
# COMPLETE
binary_vectorizer = CountVectorizer(binary=True)
binary_vectorizer.fit(df['text'].tolist())
bag_of_words_array = binary_vectorizer.transform(df['text']).toarray()

data = df.copy()
data["bag_of_words"] = list(bag_of_words_array)
data.head(5)
```

```python
frequency_vectorizer = CountVectorizer(binary=False)
frequency_vectorizer.fit(data['text'].tolist())
bag_of_words_frequency_array = frequency_vectorizer.transform(data['text']).toarray()

data["bag_of_words_frequency"] = list(bag_of_words_frequency_array)
data.head(5)
```

## Compute cosine similarities

```python
# COMPLETE
bigram_vectorizer = CountVectorizer(ngram_range=(2, 2))
bigram_features = bigram_vectorizer.fit_transform(data['text']).toarray()

data["word_bigrams"] = list(bigram_features)
data.head(5)
```

```python
char_trigram_vectorizer = CountVectorizer(analyzer='char_wb', ngram_range=(3, 3))
char_trigram_features = char_trigram_vectorizer.fit_transform(data['text']).toarray()

data["char_trigrams"] = list(char_trigram_features)
data.head(5)
```

```python
tfidf_vectorizer = TfidfVectorizer(norm='l2')
tfidf_features = tfidf_vectorizer.fit_transform(data['text']).toarray()

data["tfidf_features"] = list(tfidf_features)
data.head(5)
```

```python
lsa_50_transformer = TruncatedSVD(n_components=50)
lsa_50_features = lsa_50_transformer.fit_transform(tfidf_features)

data["lsa_50_features"] = list(lsa_50_features)
data.head(5)
```

```python
lsa_100_transformer = TruncatedSVD(n_components=100)
lsa_100_features = lsa_100_transformer.fit_transform(tfidf_features)

data["lsa_100_features"] = list(lsa_100_features)
data.head(5)
```

```python
# COMPLETE
result = {}

def calculate_max_similarity(embeddings):
    similarities = cosine_similarity(embeddings)
    max_sim = 0
    index_pair = (None, None)
    for i in range(len(similarities)):
        for j in range(len(similarities[0])):
            if i != j and similarities[i][j] > max_sim:
                max_sim = similarities[i][j]
                index_pair = (i, j)
    return index_pair, max_sim

def obtain_tweets(feature, result):
    result[feature] = {}
    for tweet_label in ["YES", "NO"]:
        filtered_data = data[data["label"] == tweet_label]
        embedding_vectors = np.array(filtered_data[feature].tolist())
        index_pair, max_similarity = calculate_max_similarity(embedding_vectors)
        result[feature][tweet_label] = {
            "tweets": [
                filtered_data["text"].iloc[index_pair[0]],
                filtered_data["text"].iloc[index_pair[1]]
            ],
            "max": round(max_similarity, 4)
        }
    return result

for feature in ['bag_of_words', 'bag_of_words_frequency', 'word_bigrams', 'char_trigrams', 'tfidf_features', 'lsa_50_features', 'lsa_100_features']:
    result = obtain_tweets(feature, result)
```

## Show results

```python
# COMPLETE
for label in result:
    print(label)
    print("="*20)
    for s in result[label]:
        print(f"label:{s}")
        tweets = result[label][s]["tweets"]
        print(f"sentence1: {tweets[0]}")
        print("-"*20)
        print(f"sentence2: {tweets[1]}")
        dist = result[label][s]["max"]
        print(f"distance:{dist}")
        print("\n")
```
